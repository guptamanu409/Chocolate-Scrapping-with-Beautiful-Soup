```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

webpage = requests.get('https://s3.amazonaws.com/codecademy-content/courses/beautifulsoup/cacao/index.html')
print(webpage)

web = webpage.content
soup = bs(web,'html.parser')
print(soup)
print(soup.find_all(attrs={'class':'Rating'}))
print(type(soup.find_all(attrs={'class':'Rating'})))
rating = soup.find_all(attrs={'class':'Rating'})
print(rating[0:5])
ratings = []
for rate in rating[1:]:
  ratings.append(float(rate.string))
print(ratings)

#Create a Histogram of Rating Values

plt.hist(ratings)
plt.show()


# Which chocolatier makes the best chocolate?

company = soup.select(".Company")
print(company[:4])
comp = []
for co in company[1:]:
  comp.append(co.string)
print('\n',comp)

dict = {'Company': comp,'Rating': ratings}
df = pd.DataFrame.from_dict(dict)
#print(df.head())

avg_ratings = df.groupby('Company').Rating.mean()
#print(avg_ratings.head())
top_ten = avg_ratings.nlargest(10)
#print(top_ten)

# Is more cacao better?

cocoa_percents = []
cocoa_percent_tags = soup.select(".CocoaPercent")
print(cocoa_percent_tags[:4])
for td in cocoa_percent_tags[1:]:
  percent = int(float(td.get_text().strip('\%')))
  cocoa_percents.append(percent)

print('\n',cocoa_percents[:3],'\n')
df['CocoaPercentage'] = cocoa_percents
print(df.head())

# Making a Scatter Plot
plt.scatter(df.CocoaPercentage, df.Rating)
# Is there any correlation here?
z = np.polyfit(df.CocoaPercentage, df.Rating, 1)
line_function = np.poly1d(z)
plt.plot(df.CocoaPercentage, line_function(df.CocoaPercentage), "r--")
plt.show()
```
